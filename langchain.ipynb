```python
## prompt templates

from langchain.prompts import PromptTemplate

promt_template = PromptTemplate(input_variables=['country'],
               template="Tell me the capital of this {country}")


promt_template.format(country="India")
```

```python
llm.predict(promt_template= promt_template, text=["India"])
```

```python
from langchain.chains import LLMChain

chain = LLMChain(llm= llm, prompt=promt_template)
chain.run("India")
```

```python
# Combining multiple chains using simple sequential chain
```

```python
capital_prompt = PromptTemplate(input_variables=["Country"],
                                template="Please tell ne the capital of the {country}")


capital_chain = LLMChain(llm = llm, prompt=capital_prompt)

famous_template = PromptTemplate(input_variables=["capital"], 
                                                  template="Suggest me some amazing places to visit in {capital}")
```

```python
famous_chain = LLMChain(llm = llm, prompt=famous_template)
```

```python
from langchain.chains import SimpleSequentialChain
```

```python
chain = SimpleSequentialChain(chains=[capital_chain, famous_chain])
chain.run("India")
```

```python
## Prompt Templates + LLM + output parser
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts.chat import ChatPromptTemplate
from langchain.schema import BaseOutputParser
```

```python
class commaseperatedoutput(BaseOutputParser):
    def parse(self, text: str):
        return text.strip().split(",")
```

```python
template = "you are a helpful assistant. When the user given any input, you should generate 5 words in comma seperated"

human_template = "{text}"
chatpromt = ChatPromptTemplate.from_messages([
    ('system', template),
    ("human", human_template)
])
```

```python
chain = chatpromt|chatllm|commaseperatedoutput()
```

```python
chain.invoke({"text":"intelligent"})
```
